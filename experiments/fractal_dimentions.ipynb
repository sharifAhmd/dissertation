## Fractal Dimention with Box Plot

```python
## this is for getting the values for max of a range
class FractalDimention:

    def __init__(self, size_marker, sizes, days):
        self.size_marker = size_marker
        self.sizes = sizes
        self.days = days
    
    def box_counting(self, ts):
        """
        Returns the box counts for a time series using given box sizes.
        """
        counts = []
        for size in self.sizes:
            boxes = set()
            for i in range(len(ts)):
                box_x = i // size
                #print(i, ts[i] // size)
                box_y = int(ts[i] // size)
                boxes.add((box_x, box_y))
            counts.append(len(boxes))
        return counts
    
    def calculate_fractal_dim(self, ts):
        """
        Returns the fractal dimension of a time series.
        """
        counts = self.box_counting(ts)

        ## track box counts
        box_counts.append(counts)

        ## Compute the log values
        log_sizes = np.log(1.0 / self.sizes)
        log_counts = np.log(counts)

        ## Compute the fractal dimension as the slope of the log-log plot
        coeffs = np.polyfit(log_sizes, log_counts, 1)
        return coeffs[0]

    
    def process_data(self):
        """
        Calculate fractal dimentions in different scales..
        """
        
        total_rows = self.days*24*60
        len_df = len(df_all)
        intervals = [(i, i+total_rows) for i in range(0, len_df, total_rows)]

        sample_taken_times, frac_dims, max_values = [], [], []

        for i in range(len(intervals)):
            st, end = intervals[i]
            df_selected = df_all[st:end]
            sample_taken_time = df_selected["time_tag"].tolist()[0].split(" ")[0]
            sample_taken_times.append(sample_taken_time)
            ##collect data for given days
            max_values.append(max(df_selected["xs"].tolist()))        
        return sample_taken_times, frac_dims, max_values
```

```python
class FractalDimention:

    def __init__(self, size_marker, sizes, days):
        self.size_marker = size_marker
        self.sizes = sizes
        self.days = days
    
    def box_counting(self, ts):
        """
        Returns the box counts for a time series using given box sizes.
        """
        counts = []
        for size in self.sizes:
            boxes = set()
            for i in range(len(ts)):
                box_x = i // size
                box_y = int(ts[i] // size)
                boxes.add((box_x, box_y))
            counts.append(len(boxes))
        return counts
    
    def calculate_fractal_dim(self, ts):
        """
        Returns the fractal dimension of a time series.
        """
        counts = self.box_counting(ts)

        ## track box counts
        box_counts.append(counts)

        ## Compute the log values
        log_sizes = np.log(1.0 / self.sizes)
        log_counts = np.log(counts)

        ## Compute the fractal dimension as the slope of the log-log plot
        coeffs = np.polyfit(log_sizes, log_counts, 1)
        return coeffs[0]

    
    def process_data(self):
        """
        Calculate fractal dimentions in different scales..
        """
        
        total_rows = self.days*24*60
        len_df = len(df_all)
        intervals = [(i, i+total_rows) for i in range(0, len_df, total_rows)]

        sample_taken_times, frac_dims = [], []

        for i in range(len(intervals)):
            st, end = intervals[i]
            df_selected = df_all[st:end]
            sample_taken_time = df_selected["time_tag"].tolist()[0].split(" ")[0]
            sample_taken_times.append(sample_taken_time)
            ##collect data for given days
            frac_dim = self.calculate_fractal_dim(df_selected["log_xs"].tolist())
            frac_dims.append(frac_dim)

            print("Sample time:", sample_taken_time, " Fractal Dimention:",frac_dim)
        
        return sample_taken_times, frac_dims
```

```python
def default_int64_handler(o):
    if isinstance(o, np.int64):
        return int(o)
    raise TypeError("Object of type '{}' is not JSON serializable".format(type(o).__name__))
```

```python
box_counts = []
size_marker = "C"
print(sizes)
scaling = "log"
days_interval = [7, 150, 365]
for days in days_interval:
    box_counts = []
    obj = FractalDimention(size_marker, sizes, days)
    sample_taken_times, frac_dims = obj.process_data()

    ## Data collections
    collect_data = {}
    
    collect_data["min_value"] = min_value
    collect_data["max_value"] = max_value
    collect_data[f"scaling_{scaling}_sizes"] = list(sizes)
    collect_data[f"scaling_{scaling}_box_counts"] = box_counts
    collect_data[f"scaling_{scaling}_sample_times"] = sample_taken_times
    collect_data[f"scaling_{scaling}_FD"] = frac_dims

    import json
    with open(f"FD_results/type_{size_marker}_scaling_{scaling}_days_{days}.json", 'w') as json_file:
        json.dump(collect_data, json_file, default=default_int64_handler)
```
